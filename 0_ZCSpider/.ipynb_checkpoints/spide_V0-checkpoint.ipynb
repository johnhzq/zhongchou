```python
def getOnePageList(url):    
    html = requests.get(url)
    html.encoding = html.apparent_encoding
    text_html = html.text
    
    bsObj = BeautifulSoup(text_html, "html.parser")    
    ItemBsObjs = bsObj.find_all("div", {"class": "ssCardItem"})
    
    rsDF = pd.DataFrame()    
    for Item in ItemBsObjs:
        blankDF = pd.DataFrame(np.zeros([1,7]),columns=['title','url','brief','tags','ft_v','sc_v','th_v'])
        ############
        ssCardICText = Item.find("div", {"class": "ssCardICText"})
        a_title = ssCardICText.find('a')
        title = a_title["title"]
        url = a_title["href"]
        brief = ssCardICText.find('p').text
        ############
        siteCardFLabelBox_siteIlB_box = Item.find("div", {"class": "siteCardFLabelBox siteIlB_box"})
        siteIlB_items = siteCardFLabelBox_siteIlB_box.find_all('a')
        tags = []
        for item in siteIlB_items:
            tags.append(item.text.strip())   
        ############
        siteCardFData = Item.find("div", {"class": "siteCardFData"})
        ft = siteCardFData.find("div", {"class": "ftDiv"})
        ft_v = int(ft.find('p').text.strip().replace('￥',''))
        sc = siteCardFData.find("div", {"class": "scDiv"})
        sc_v = int(sc.find('p').text.strip())
        th = siteCardFData.find("div", {"class": "thDiv"})
        th_v = int(th.find('p').text.strip().replace('%',''))
        ########################
        blankDF.title = title.strip().replace('|','')
        blankDF.url = url.strip().replace('|','')
        blankDF.brief = brief.strip().replace('\r\n','')
        blankDF.tags = '#'.join(tags)
        blankDF.ft_v = ft_v
        blankDF.sc_v = sc_v
        blankDF.th_v = th_v
        rsDF = pd.concat([rsDF, blankDF], ignore_index=True, axis=0)
    return rsDF
```

```python
def getCount(_id):
#     sleep(3)
    progressListURL = 'http://www.zhongchou.com/deal-march_list?id='+_id
    print('【getCount】',progressListURL)
    headers = {'User-Agent': getUA()}
    proxies = { 'http' : getPROXY()} 
    try:
        html = requests.get(progressListURL, headers=headers, proxies=proxies, timeout=10)
        html.encoding = html.apparent_encoding
        text_html = html.text
        jsonData = json.loads(text_html)
        return jsonData['data']['count']
    except:
        pass
        return None
    
def getProgressList_json(_id, page_size):
#     sleep(3)
    if int(page_size)>0:
        progressListURL = 'http://www.zhongchou.com/deal-march_list?id='+_id+'&page_size='+page_size
    else:
        progressListURL = 'http://www.zhongchou.com/deal-march_list?id='+_id
    print('【getProgressList_json】',progressListURL)
    headers = {'User-Agent': getUA()}
    proxies = { 'http' : getPROXY()}   
    try:
        html = requests.get(progressListURL, headers=headers, proxies=proxies, timeout=10)        
        html.encoding = html.apparent_encoding
        text_html = html.text
        jsonData = json.loads(text_html)
        return jsonData
    except:
        pass
        return None

def getProgressList(_id):
    rs = None
    while True:
        page_size = getCount(_id)
        if page_size is None:
            continue
        break        
    print('【getProgressList】',page_size)
    while True:
        jsonData = getProgressList_json(_id, page_size)
        if jsonData is None:
            continue
        if jsonData['status']==1:
            rs = list()
            data = jsonData['data']
            for one in data['march_list']:
    #             print(one['create_time'],one['user_name'],one['log_info'])
                line = []
                line.append(one['create_time'])
                line.append(one['user_name'])
                line.append(one['log_info'])
                rs.append('-'.join(line))
    #     else:
    #         print('请求数据失败！！！')
            break
    return rs
```

```python
def getXQ_ZC(_id):
    url = 'http://www.zhongchou.com/deal-show/id-'+_id
    html = requests.get(url)
    html.encoding = html.apparent_encoding
    text_html = html.text
    bsObj = BeautifulSoup(text_html, "html.parser")
    ###################### 【项目介绍】
    xmxqBox_BsObj = bsObj.find("div", {"id": "xmxqBox"})
#     print('#'*25+'【项目介绍】'+'#'*25)
    xq = xmxqBox_BsObj.text.strip().replace('	','').replace(' ','').replace('\xa0','')
    # print(xq)
    ###################### 【回报项】
    zcjeOuterBox_BsObj = bsObj.find("div", {"class": "zcjeOuterBox"})
    zcje_ItemBox_BsObj = zcjeOuterBox_BsObj.find_all("div", {"class": "zcje_ItemBox"})
#     print('#'*25+'【回报项】'+'#'*25)
    zc = []
    for one in zcje_ItemBox_BsObj:
        item = []
        zcje = one.find('b').text
        zcrs = one.find('h3').text.replace(zcje,'')
        text = one.text.replace(zcje,'').replace(zcrs,'').strip().replace('\r\n','').replace('\n','')
        item.append(zcje)
        item.append(zcrs)
        item.append(text)
        zc.append('-'.join(item))
#     print(zc)
    return xq,zc
```

```python
def getOneProjectData(url):
    _id = url.split('-')[-1]
    tryNum=1
    while tryNum<=3:
        gx = getProgressList(_id) # 项目更新（gx）
        if gx is None:
            print('【getOneProjectData】 《getProgressList》 第'+str(tryNum)+'次请求数据失败！！！')
        else:
            break
        tryNum+=1
    xq,zc = getXQ_ZC(_id) # 项目详情（xq）、支持，几回报项（zc）
    return '\n'.join(gx),xq,'\n'.join(zc)
```
