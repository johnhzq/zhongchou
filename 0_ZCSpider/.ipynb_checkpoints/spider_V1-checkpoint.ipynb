```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np

def getOnePageList(url):    
    html = requests.get(url)
    html.encoding = html.apparent_encoding
    text_html = html.text
    
    bsObj = BeautifulSoup(text_html, "html.parser")    
    ItemBsObjs = bsObj.find_all("div", {"class": "ssCardItem"})
    
    rsDF = pd.DataFrame()    
    for Item in ItemBsObjs:
        blankDF = pd.DataFrame(np.zeros([1,7]),columns=['title','url','brief','tags','ft_v','sc_v','th_v'])
        ############
        ssCardICText = Item.find("div", {"class": "ssCardICText"})
        a_title = ssCardICText.find('a')
        title = a_title["title"]
        url = a_title["href"]
        brief = ssCardICText.find('p').text
        ############
        siteCardFLabelBox_siteIlB_box = Item.find("div", {"class": "siteCardFLabelBox siteIlB_box"})
        siteIlB_items = siteCardFLabelBox_siteIlB_box.find_all('a')
        tags = []
        for item in siteIlB_items:
            tags.append(item.text.strip())   
        ############
        siteCardFData = Item.find("div", {"class": "siteCardFData"})
        ft = siteCardFData.find("div", {"class": "ftDiv"})
        ft_v = int(ft.find('p').text.strip().replace('￥',''))
        sc = siteCardFData.find("div", {"class": "scDiv"})
        sc_v = int(sc.find('p').text.strip())
        th = siteCardFData.find("div", {"class": "thDiv"})
        th_v = int(th.find('p').text.strip().replace('%',''))
        ########################
        blankDF.title = title.strip().replace('|','')
        blankDF.url = url.strip().replace('|','')
        blankDF.brief = brief.strip().replace('\r\n','')
        blankDF.tags = '#'.join(tags)
        blankDF.ft_v = ft_v
        blankDF.sc_v = sc_v
        blankDF.th_v = th_v
        rsDF = pd.concat([rsDF, blankDF], ignore_index=True, axis=0)
    return rsDF
```

```python
import requests
import json

def getSupportList_Count_ET(_id):
    count = None
    ET = None
    supportListURL = 'http://www.zhongchou.com/deal-support_list?id='+_id
    print('【Count_ET】',supportListURL)
    headers = {'User-Agent': getUA()}
    proxies = { 'http' : getPROXY()} 
    try:
        html = requests.get(supportListURL, headers=headers, proxies=proxies, timeout=10)
        html.encoding = html.apparent_encoding
        text_html = html.text
        jsonData = json.loads(text_html)
        count = jsonData['data']['count']
        if len(jsonData['data']['support_list'])>0:
            ET = jsonData['data']['support_list'][0]['pay_time']
    except Exception as e:
        print(e)
    return count,ET

def getSupportList_ST(_id,page_size):
    ST = None
    if int(page_size)>0:
        supportListURL = 'http://www.zhongchou.com/deal-support_list?id='+_id+'&page_size='+page_size
    else:
        supportListURL = 'http://www.zhongchou.com/deal-support_list?id='+_id
    print('【ST】',supportListURL)
    headers = {'User-Agent': getUA()}
    proxies = { 'http' : getPROXY()} 
    try:
        html = requests.get(supportListURL, headers=headers, proxies=proxies, timeout=10)
        html.encoding = html.apparent_encoding
        text_html = html.text
        jsonData = json.loads(text_html)
        if len(jsonData['data']['support_list'])>0:
            ST = jsonData['data']['support_list'][-1]['pay_time']
    except Exception as e:
        print(e)
    return ST
```

# 3、项目更新

```python
import requests
import json
from time import sleep

def getCount(_id):
    count = None
    progressListURL = 'http://www.zhongchou.com/deal-march_list?id='+_id
    print('【getCount】',progressListURL)
    headers = {'User-Agent': getUA()}
    proxies = { 'http' : getPROXY()} 
    try:
        html = requests.get(progressListURL, headers=headers, proxies=proxies, timeout=10)
        html.encoding = html.apparent_encoding
        text_html = html.text
        jsonData = json.loads(text_html)
        count = jsonData['data']['count']
    except Exception as e:
        print(e)
    return count
    
def getProgressList_json(_id, page_size):
    progressList_json = None
    if int(page_size)>0:
        progressListURL = 'http://www.zhongchou.com/deal-march_list?id='+_id+'&page_size='+page_size
    else:
        progressListURL = 'http://www.zhongchou.com/deal-march_list?id='+_id
    print('【getProgressList_json】',progressListURL)
    headers = {'User-Agent': getUA()}
    proxies = { 'http' : getPROXY()}   
    try:
        html = requests.get(progressListURL, headers=headers, proxies=proxies, timeout=10)        
        html.encoding = html.apparent_encoding
        text_html = html.text
        progressList_json = json.loads(text_html)
    except Exception as e:
        print(e)
    return progressList_json
##################################
def getProgressList(_id):
    rs = None
    while True:
        page_size = getCount(_id)
        if page_size is None:
            sleep(3)
            continue
        break        
    print('【getProgressList】',page_size)
    while True:
        jsonData = getProgressList_json(_id, page_size)
        if jsonData is None:
            sleep(3)
            continue
        if jsonData['status']==1:
            rs = list()
            data = jsonData['data']
            for one in data['march_list']:
    #             print(one['create_time'],one['user_name'],one['log_info'])
                line = []
                line.append(one['create_time'].replace('|',''))
                line.append(one['user_name'].replace('|',''))
                line.append(one['log_info'].replace('|',''))
                rs.append('-'.join(line))
    #     else:
    #         print('请求数据失败！！！')
            break
    return rs
```

# 4、项目详情-回报项-tags

#### tags
- gy siteIlB_item
- addr siteIlB_item
- label siteIlB_item

```python
import requests
from bs4 import BeautifulSoup

def getXQ_ZC_tags(_id):
    url = 'http://www.zhongchou.com/deal-show/id-'+_id
    html = requests.get(url)
    html.encoding = html.apparent_encoding
    text_html = html.text
    bsObj = BeautifulSoup(text_html, "html.parser")
    ###################### 【项目介绍】
    xmxqBox_BsObj = bsObj.find("div", {"id": "xmxqBox"})
#     print('#'*25+'【项目介绍】'+'#'*25)
    xq = xmxqBox_BsObj.text.strip().replace('	','').replace(' ','').replace('\xa0','').replace('|','')
    # print(xq)
    ###################### 【回报项】
    zcjeOuterBox_BsObj = bsObj.find("div", {"class": "zcjeOuterBox"})
    zcje_ItemBox_BsObj = zcjeOuterBox_BsObj.find_all("div", {"class": "zcje_ItemBox"})
#     print('#'*25+'【回报项】'+'#'*25)
    zc = []
    for one in zcje_ItemBox_BsObj:
        item = []
        zcje = one.find('b').text
        zcrs = one.find('h3').text.replace(zcje,'')
        text = one.text.replace(zcje,'').replace(zcrs,'').strip().replace('\r\n','').replace('\n','').replace('|','')
        item.append(zcje)
        item.append(zcrs)
        item.append(text)
        zc.append('-'.join(item))
#     print(zc)
    ###################### 【tags】    
    tages_BsObj = bsObj.find("div", {"class": "jlxqTitleText siteIlB_box"})
    yg=None
    yg_BsObj = tages_BsObj.find("span", {"class": "gy siteIlB_item"})
    if not yg_BsObj is None:
        yg = yg_BsObj.text.strip().replace('\n','#')        
    addr=None
    addr_BsObj = tages_BsObj.find("span", {"class": "addr siteIlB_item"})
    if not addr_BsObj is None:
        addr = addr_BsObj.text.strip().replace('\n','#') 
    label=None
    label_BsObj = tages_BsObj.find("span", {"class": "label siteIlB_item"})
    if not label_BsObj is None:
        label = label_BsObj.text.strip().replace('\n','#') 
#     print(yg,addr,label)
    
    return xq,zc,yg,addr,label
```

# 5、OneProjectData

```python
from time import sleep
def getOneProjectData(url):
    _id = url.split('-')[-1]
    ST,ET = getST_ET(_id) # 时间
    gx = getProgressList(_id) # 项目更新（gx）       
    xq,zc,yg,addr,label = getXQ_ZC_tags(_id) # 项目详情（xq）、支持，几回报项（zc）
    ############################################
    one = []
    one.append('\n'.join(gx))
    one.append(xq)
    one.append('\n'.join(zc))
    doc = '\n'.join(one).replace('\r\n','').replace('\n','#').replace('|','').replace(' ','')
    print('【getOneProjectData】 len(doc)=',len(doc))
    
    return ST,ET,doc,yg,addr,label
```
