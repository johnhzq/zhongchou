```python
inPutDir = 'inPut/'
outPutDir = 'outPut/'
```

```python
from MyUtils import MyUtils
import pandas as pd

startPage = 1
endPage = 50
myUtils = MyUtils()
##################################################################
data0 = pd.DataFrame()
while startPage<=endPage: 
    #######################################################
    pageNum = startPage
#     print('#'*30+'【pageNum ='+str(pageNum)+'】'+'#'*30)
    #######################################################
    TXT = inPutDir+'DOC_Page_'+str(pageNum)+'.txt'
    seq = '|'
    onePage = myUtils.TXT2DF(TXT,seq)
    if onePage.shape[0]>24:
        print(onePage.shape)
    data0 = pd.concat([data0, onePage], ignore_index=True, axis=0)
    #######################################################
    startPage+=1
data0 = data0.astype(str)
print(data0.shape)
data0.head()
```

```python
data0['len_doc']=data0['len_doc'].astype(int)
data1 = data0[data0.len_doc>=500].reset_index(drop=True)
print(data1.shape)
data1.head(3)
```

```python
data2 = data1[['title','url','addr','label']]
print(data2.shape)
data2.head(3)
```

```python
data2['province'] = ""
data2['city'] = ""
data2['product'] = ""
data2.head()
```

```python
rows = data2.shape[0]
number = rows//4 if rows%4==0 else rows//4+1 # = 285
for i in range(4):
    start = i*number
    end = (i+1)*number
    onepart = data2[(data2.index>=start)&(data2.index<end)]
#     print(onepart.shape)
    onepart.to_excel(outPutDir+"part_"+str(start)+"_"+str(end)+".xlsx",encoding='utf8')
```
