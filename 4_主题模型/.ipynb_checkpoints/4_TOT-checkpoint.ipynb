# 2、TOT

### 2.1 GetCorpusAndDictionary
return documents, timestamps, dictionary

```python
def GetCorpusAndDictionary(data):
    ## documents
    documents = list(data.doc.apply(lambda doc: doc.split()))
    ## dictionary
    dictionary = set()
    for words in documents:
        dictionary.update(words)
    dictionary = list(dictionary)
    ## timestamps
    first_timestamp = data.time.max()
    last_timestamp = data.time.min()
    timestamps = list(data.time.apply(lambda t: 1.0*(t-first_timestamp)/(last_timestamp-first_timestamp)))
    assert len(documents) == len(timestamps),'文档数量与时间戳数量不等！！！'
    return documents, timestamps, dictionary
```

### 2.2 InitializeParameters
return par

```python
def CalculateCounts(par):
    for d in range(par['D']):
        for i in range(par['N'][d]):
            topic_di = par['z'][d][i]				# 第d个文本中，第i个单词属于第几个主题
            word_di = par['w'][d][i]				# 第d个文本中，第i个单词的ID（即dictionary中的位置）
            par['m'][d][topic_di] += 1				# 文本-主题矩阵中第d个文本的topic_di对应的主题数量加一
            par['n'][topic_di][word_di] += 1		# 主题-单词矩阵中topic_di对应的主题中的word_di对应的单词数量加一
            par['n_sum'][topic_di] += 1				# topic_di对应的主题的计数加一(每个主题下的单词数)

import random
import scipy.special
import numpy as np

def InitializeParameters(documents, timestamps, dictionary, max_iterations=10, T=10):
    alpha = 50/T+1 			# alpha是一个对称的Dirichlet分布的参数，值越大意味着越平滑（更正规化）
    beta = 0.1 	 	 	 	#
    psi = 1 				# 即初始时，假设 t 时间所有主题均出现的概率密度函数的值均为 1
                            # t = 0.3
                            # f(t;1,1) = math.pow(t, 1-1) * math.pow(1-t, 1-1) / scipy.special.beta(1,1) = 1
    
    par = {}														# 所有参数的字典
    par['dataset'] = 'dxfp'											# 数据集名称
    par['max_iterations'] = max_iterations										# Gibbs抽样的最大迭代次数
    par['T'] = T													# 主题总数
    par['D'] = len(documents)										# 文本总数
    par['V'] = len(dictionary)										# 单词总数
    par['N'] = [len(doc) for doc in documents]						#每篇文本的单词数量 
    par['alpha'] = [alpha  for _ in range(par['T'])]			# α 阿尔法
    par['beta'] = [beta for _ in range(par['V'])]					# β 贝塔
    par['beta_sum'] = sum(par['beta'])								# β 和
    par['psi'] = [[psi for _ in range(2)] for _ in range(par['T'])]	# Ψ 普西
    par['betafunc_psi'] = [scipy.special.beta( par['psi'][t][0], par['psi'][t][1] ) for t in range(par['T'])]
    # 词的字典
    par['word_id'] = {dictionary[i]: i for i in range(len(dictionary))}
    # 词条
    par['word_token'] = dictionary
    # 为文本中的每个单词随机分配一个主题
    par['z'] = [[random.randrange(0,par['T']) for _ in range(par['N'][d])] for d in range(par['D'])]
    # 为文本中的每个单词添加一个时间标签，同一篇文章中的所有单词都是同一个时间标签
    par['t'] = [[timestamps[d] for _ in range(par['N'][d])] for d in range(par['D'])]
    # 【文档-词】
    par['w'] = [[par['word_id'][documents[d][i]] for i in range(par['N'][d])] for d in range(par['D'])]
    # 【文本-主题】
    par['m'] = [[0 for t in range(par['T'])] for d in range(par['D'])]
    # 【主题-词】
    par['n'] = [[0 for v in range(par['V'])] for t in range(par['T'])]
    # 主题计数
    par['n_sum'] = [0 for t in range(par['T'])]
    # 设置打印选项：打印数组所有元素
    np.set_printoptions(threshold=np.inf)
    # 设置如何处理浮点错误：
    np.seterr(divide='ignore', invalid='ignore')
    CalculateCounts(par)
    return par
```

### 2.3 GetMethodOfMomentsEstimatesForPsi
return psi

```python
# list(filter(lambda x: x!=0, [1,2,3,4,0,0,1,0,0])) # [1, 2, 3, 4, 1]
# tt = []
# tt.extend([1,2])
# tt.extend([2,3])
# tt # [1, 2, 2, 3]
```

```python
def GetTopicTimestamps(par):		# 获取主题时间戳
    topic_timestamps = [] 
    for topic in range(par['T']):
        current_topic_timestamps = []
        current_topic_doc_timestamps = [[ (par['z'][d][i]==topic)*par['t'][d][i] for i in range(par['N'][d])] for d in range(par['D'])]
        for d in range(par['D']):
            current_topic_doc_timestamps[d] = list(filter(lambda x: x!=0, current_topic_doc_timestamps[d]))
        for timestamps in current_topic_doc_timestamps:
            current_topic_timestamps.extend(timestamps)
        assert current_topic_timestamps != []
        topic_timestamps.append(current_topic_timestamps)
    return topic_timestamps

# PSI的矩估计方法 
def GetMethodOfMomentsEstimatesForPsi(par):
    topic_timestamps = GetTopicTimestamps(par)
    psi = [[1 for _ in range(2)] for _ in range(len(topic_timestamps))]
    for i in range(len(topic_timestamps)):
        current_topic_timestamps = topic_timestamps[i]
        timestamp_mean = np.mean(current_topic_timestamps)		# 均值
        timestamp_var = np.var(current_topic_timestamps)		# 方差
        if timestamp_var == 0:
            timestamp_var = 1e-6
        common_factor = timestamp_mean*(1-timestamp_mean)/timestamp_var - 1
        psi[i][0] = 1 + timestamp_mean*common_factor			# 【对比公式14，多了一个：1】
        psi[i][1] = 1 + (1-timestamp_mean)*common_factor		# 【对比公式14，多了一个：1】

    return psi
```

### 2.4 ComputePosteriorEstimatesOfThetaAndPhi
return theta, phi

```python
import copy

# Theta 和 Phi 的后验估计
def ComputePosteriorEstimatesOfThetaAndPhi(par):
    theta = copy.deepcopy(par['m']) # 【文档-主题】
    phi = copy.deepcopy(par['n']) # 【主题-词】

    for d in range(par['D']):
        if sum(theta[d]) == 0:
            theta[d] = np.asarray([1.0/len(theta[d]) for _ in range(len(theta[d]))])
        else:
            theta[d] = np.asarray(theta[d])
            theta[d] = 1.0*theta[d]/sum(theta[d])
    theta = np.asarray(theta) # 【文档-主题】每个文档下不同主题出现的概率

    for t in range(par['T']):
        if sum(phi[t]) == 0:
            phi[t] = np.asarray([1.0/len(phi[t]) for _ in range(len(phi[t]))])
        else:
            phi[t] = np.asarray(phi[t])
            phi[t] = 1.0*phi[t]/sum(phi[t])
    phi = np.asarray(phi) # 【主题-词】每个主题下不同词出现的概率

    return theta, phi

# θ的后验估计
def ComputePosteriorEstimatesOfTheta(self, par):
    theta = copy.deepcopy(par['m'])

    for d in range(par['D']):
        if sum(theta[d]) == 0:
            theta[d] = np.asarray([1.0/len(theta[d]) for _ in range(len(theta[d]))])
        else:
            theta[d] = np.asarray(theta[d])
            theta[d] = 1.0*theta[d]/sum(theta[d])

    return np.matrix(theta)

# φ的后验估计
def ComputePosteriorEstimateOfPhi(self, par):
    phi = copy.deepcopy(par['n'])

    for t in range(par['T']):
        if sum(phi[t]) == 0:
            phi[t] = np.asarray([1.0/len(phi[t]) for _ in range(len(phi[t]))])
        else:
            phi[t] = np.asarray(phi[t])
            phi[t] = 1.0*phi[t]/sum(phi[t])

    return np.matrix(phi)
```

### 2.5 TopicsOverTimeGibbsSampling
return par['m'], par['n'], par['psi']

```python
def TopicsOverTimeGibbsSampling(par):
    for iteration in range(par['max_iterations']):
        print('Done with iteration {iteration}'.format(iteration=iteration))
        for d in range(par['D']):
            for i in range(par['N'][d]):
                word_di = par['w'][d][i]
                t_di = par['t'][d][i]
                #// for the current assignment of k to the word token wm,n:
                old_topic = par['z'][d][i]
                par['m'][d][old_topic] -= 1
                par['n'][old_topic][word_di] -= 1
                par['n_sum'][old_topic] -= 1
                #// multinomial sampling according to Eq. (8):
                topic_probabilities = []
                for topic_di in range(par['T']):
                    psi_di = par['psi'][topic_di]
                    topic_probability = 1.0 * (par['m'][d][topic_di] + par['alpha'][topic_di])
                    topic_probability *= ((1-t_di)**(psi_di[0]-1)) * ((t_di)**(psi_di[1]-1))
                    topic_probability /= par['betafunc_psi'][topic_di]
                    topic_probability *= (par['n'][topic_di][word_di] + par['beta'][word_di])
                    topic_probability /= (par['n_sum'][topic_di] + par['beta_sum'])                    
                    topic_probabilities.append(topic_probability)
                sum_topic_probabilities = sum(topic_probabilities)
                if sum_topic_probabilities is np.nan:
                        print('【',d,'-',i,'】',topic_probabilities) 
                if sum_topic_probabilities == 0:
                    topic_probabilities = [1.0/par['T'] for _ in range(par['T'])]
                else:
                    topic_probabilities = [p/sum_topic_probabilities for p in topic_probabilities]                 
#                 new_topic = list(np.random.multinomial(1, topic_probabilities, size=1)[0]).index(1)
                temp_new_topic = list(np.random.multinomial(1, topic_probabilities, size=1)[0])
                if 1 not in temp_new_topic:
                    print(temp_new_topic)
                new_topic = temp_new_topic.index(1)
                #// for the new assignment of zm,n to the word token wm,n:
                par['z'][d][i] = new_topic
                par['m'][d][new_topic] += 1
                par['n'][new_topic][word_di] += 1
                par['n_sum'][new_topic] += 1
            #
#             if d%100 == 0:
#                 print('Done with iteration {iteration} and document {document}'.format(iteration=iteration, document=d))
        #// update Ψ after each Gibbs sample by the method of moments
        par['psi'] = GetMethodOfMomentsEstimatesForPsi(par)
        par['betafunc_psi'] = [scipy.special.beta( par['psi'][t][0], par['psi'][t][1] ) for t in range(par['T'])] 
    # read out parameter set Φ according to Eq. (11)
    # read out parameter set Θ according to Eq. (12)
    par['m'], par['n'] = ComputePosteriorEstimatesOfThetaAndPhi(par)
    return par['m'], par['n'], par['psi']
```

```python
import time
import pickle

def TOT(data,min_T,max_T,step,X,outPutDir):
    documents, timestamps, dictionary = GetCorpusAndDictionary(data)
    for T in np.arange(min_T,max_T+step,step):
        print('-'*40,T,'-'*40)
        startTime = time.time()
        ### start 
        par = InitializeParameters(documents, timestamps, dictionary, 10, T)
        theta, phi, psi = TopicsOverTimeGibbsSampling(par)
        #######################################################
        tot_pickle_path = outPutDir+'TOT_'+X+'_'+str(T)+'.pickle'    
        tot_pickle = open(tot_pickle_path,'wb')
        pickle.dump(par, tot_pickle)
        tot_pickle.close()
        ### end
        endTime = time.time()
        print(time.strftime("%Y-%m-%d %H:%M:%S", time.localtime(startTime)))
        print(time.strftime("%Y-%m-%d %H:%M:%S", time.localtime(endTime)))
        print((endTime-startTime)/60)
```

```python
# outPutDir = 'outPut/'
min_T = 10
max_T = 30
step = 5

for _type in range(1,7):
    if _type==1:
        print('#'*40,'【S】','#'*40)
        DATA = data[['time','doc']].reset_index(drop=True)
        print(DATA.head(2))
        TOT(DATA,min_T,max_T,step,'S',outPutDir)
    else:
        X = data.columns[_type]
        print('#'*40,'【'+X+'】','#'*40)
        DATA = data[data[X]==1][['time','doc']].reset_index(drop=True)        
        print(DATA.head(2))
        TOT(DATA,min_T,max_T,step,X,outPutDir)
```
